```python
#QUESTION  2 

df1 =df['processor_brand'].value_counts()
print('number of laptops of each processor brand :',df1)
print('Most available processor brand laptop :',df1.idxmax())
print('Most unavailable processor brand laptop :',df1.idxmin())
```
