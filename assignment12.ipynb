```python
                         # TYPE OF PLOTS
# 1)  RELATIONAL PLOT:
#It is basically used to show the relation between any two attributes of the dataset
#It contains numerical value on both x and y axis
#It consists of 2 types of plots scatter and line
#By default it represents the data in scatter plot
#It is represented by relplot()

#a)Scatter plot
sns.relplot(x='math score',y ='reading score',data =df)  
#We can also use scatterplot() instead of relplot()
#it appears that there might be a positive correlation between math scores and reading scores. 
#This means that students who scored higher in math also tended to score higher in reading, and vice versa
```

```python
sns.relplot(x='math score',y ='reading score',data =df,hue='race/ethnicity',style='gender',col='lunch')
#there might be positive correlations between lunch standard scores and math scores, 
#and between lunch free/reduced scores and reading scores.
```

```python
#b) Line plot

sns.relplot(x='math score',y ='reading score',data =df,hue='race/ethnicity',style='gender',col='lunch',kind='line')
#the graph appears to show the number of students who have eaten lunch in different groups, 
#categorized by lunch program (standard or free/reduced) and demographics (race/ethnicity and gender)
```

```python
#a)Bar plot
sns.catplot(y='parental level of education',hue='lunch',data=df,x='test preparation course',kind='bar')
#We can also use barplot() instead of catplot()
# the graph appears to show the percentage of students who have completed a test preparation course 
#according to their parental level of education
```

```python
#b) Violin plot: 
#The width of violin plot represents the density of data points at different values

sns.catplot(y='parental level of education',hue='lunch',data=df,x='test preparation course',kind='violin')
#the graph suggests that there might be a correlation between parental level of education and income. 
#Parents with higher levels of education tend to have higher incomes.
```

```python
sns.catplot(y='parental level of education',x='lunch',hue='gender',data=df,kind='bar')
plt.title('bargraph')   #We can give a title to the plot
plt.show()
#  the graph suggests that there might be more females with a bachelor's degree than males
```

```python
sns.catplot(y='parental level of education',x='lunch',hue='gender',data=df,kind='bar')
plt.title('bargraph')
plt.ylabel('Parent Qualification') #We can label the y axis variables
plt.xlabel('Lunch')         #We can label the x axis variables
plt.show()
# the bar graph appears to show the percentage of students with a bachelor's degree by gender
```

```python
                                      #PIE CHARTS

label =list(df['race/ethnicity'].value_counts().keys())
values=list(df['race/ethnicity'].value_counts().values)
plt.pie(values,labels=label)
plt.show()

#Group C is the most common category.
```

```python
ls = [0,0,0,0,0.4]
plt.pie(values,labels=label,wedgeprops={'ec':'k'},explode= ls,autopct='%1.2f%%',shadow=True)
plt.title('pie chart')
plt.show()
```

```python
sns.boxplot(df['math score'])  
#By analyzing the position of the median line, the width of the box, and the presence of outliers, 
#you can get a sense of how students performed on the math test and how much variation there was in the scores.
```
